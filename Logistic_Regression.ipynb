# **Data Preprocessing**

```python
features.isnull().sum()
```

```python
Age_mean1=features['Age'].mean()
Age_mean1
```

```python
Cabin_mode1=features['Cabin'].mode()[0]
Cabin_mode1
```

```python
embarked_mode=features['Embarked'].mode()[0]
embarked_mode
```

```python
features.fillna({'Age':Age_mean1,'Cabin':Cabin_mode1,'Embarked':embarked_mode},inplace=True)
```

```python
features.isnull().sum()
```

```python
features.drop(columns=['Cabin'],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
```

```python
features['Sex']=label.fit_transform(features['Sex'])
```

```python
features['Embarked']=label.fit_transform(features['Embarked'])
```

```python
features
```

```python
df1=pd.merge(features,target,left_index=True,right_index=True)
```

```python
df1
```

```python
target_new=df1['Survived']
```

```python
target_new
```

**3. Model Building:**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target_new,train_size=0.75,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
log_model=LogisticRegression()
```

```python
log_model.fit(x_train,y_train)
```

```python
y_predict=log_model.predict(x_test)
```

# **4 Model Evaluation:**

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```

```python
conf=confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(conf,annot=True,fmt='d')
```

```python
log_proba=log_model.predict_proba(x_test)[:,1]
```

```python
auc=roc_auc_score(y_test,log_proba)
auc
```

```python
import matplotlib.pyplot as plt
```

```python
fpr,tpr,thr=roc_curve(y_test,log_proba)
plt.plot(fpr,tpr,linewidth=2,color='red',label=f'AUC_Score={auc:.2f}')



plt.plot([0,1],linestyle='--',color='grey')
plt.legend(loc='best')

plt.grid()
plt.xlabel('FPR',color='red',size=15)
plt.ylabel('TPR',color='red',size=15)
plt.title('ROC-RECEIVER OPRATING CHARACTERISTICS',color='black',size=15)
```

```python
%%writefile Model_Dep_app.py
import streamlit as st
import warnings
import pickle
import pandas as pd  # Import pandas

warnings.filterwarnings('ignore')

# Load the model
model = pickle.load(open('log.pkl', 'rb'))

st.title('Model Deployment using Logistic Regression')

def user_input_parameters():
    Pclass = st.sidebar.selectbox('Pclass 1-class1 2-class2 3-class3', [1, 2, 3])
    Sex = st.sidebar.selectbox('Gender 1-Male 0-Female', [0, 1])
    Age = st.sidebar.number_input('Enter the Age', min_value=0, max_value=100, value=30)
    SibSp = st.sidebar.selectbox('Select SibSp (siblings & spouses)', [1, 0, 3, 4, 2, 5, 8])
    Parch = st.sidebar.selectbox('Select Parch (parents & children)', [0, 1, 2, 3, 4, 5, 6])
    Embarked = st.sidebar.selectbox('Embarked 0-C 1-Q 2-S', [0, 1, 2])
    data = {'Pclass': Pclass, 'Sex': Sex, 'Age': Age, 'SibSp': SibSp, 'Parch': Parch, 'Embarked': Embarked}
    features = pd.DataFrame(data, index=[0])
    return features

# Get user input
df = user_input_parameters()

# Enforce column order to match the training data
df = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']]

# Display user input data
st.write("User Input Data")
st.write(df)

# Predictions
predict = model.predict(df)
pred_prob = model.predict_proba(df)

# Display predictions
st.subheader('Predicted Output')
st.write('Yes' if pred_prob[0][1] > 0.5 else 'No')
st.subheader('Predicted Probability')
st.write(pred_prob)
```
